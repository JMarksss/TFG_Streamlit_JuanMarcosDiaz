## 4) Preprocesado - Construcción de distintos datasets

En esta sección vamos a construir el dataset final con el que trabajaremos. En primer lugar vamos a elminar las variables categóricas que no influyan sobre la variable salida. Vamos a eliminar las columnas categóricas que sean independientes de la variable respuesta, es decir, no tengan relación. Para ello utilizaremos un test Chi Cuadrado de tal manera que las variables con un p-valor mayor que 0.05 serán las que cumplan la hipótesis nula de independencia y las que eliminaremos. 

```python
cat_cols = cat.select_dtypes(include=['object', 'category']).columns.to_list()
for col in cat_cols:
    ct = pd.crosstab(cat.Cardiopatía, cat[col]).to_numpy()
    chi2, p_val , dof, ex = stats.chi2_contingency(ct)
    print(f'Variable: {col}, p-valor: {p_val}')
```

Aunque la variable Glucemia no nos será útil en los modelos de clasificación, si que podría aportar información en el caso de los modelos de regresión o clustering, por lo que finalmente vamos a decidir no eliminarla.

Para la implementación de los modelos vamos a separar la variable etiqueta del resto.

```python
Y = datos['output'] #Por comodidad, recuperamos la variable original, aunque es indiferente
X = cat.drop(['Cardiopatía'], axis=1)
```

Aunque es posible que obtengamos peores resultados, aplicaremos la codificación One-Hot a las variables categóricas restantes. El dataset original ya contenía todas las variables codificadas, pero creemos que tiene más sentido la codificación One-Hot ya que, por ejemplo en el caso del dolor torácico, no consideramos que una codificación ordinal tenga sentido.

En cuanto a las variables numéricas, construiremos distintos datasets aplicando diferentes técnicas de escalado recogidas en la memoria sobre las variables originales:
- Escalado estandar.
- Escalado min-max.
- Escalado robusto.

```python
#Importamos las librerías para el preprocesado
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.compose import make_column_selector

numeric_cols = X.select_dtypes(include=['float64', 'int']).columns.to_list()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.to_list()

#Construimos los preprocesadores 
standard = ColumnTransformer(
                   [('standard', StandardScaler(), numeric_cols),
                    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'), cat_cols)],
                remainder='passthrough')

minmax = ColumnTransformer(
                   [('min-max', MinMaxScaler(), numeric_cols),
                    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'), cat_cols)],
                remainder='passthrough')

robust = ColumnTransformer(
                   [('robust', RobustScaler(), numeric_cols),
                    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'), cat_cols)],
                remainder='passthrough')
```

#### Escalado Estandar

```python
X_st = standard.fit_transform(X)
```

#### Escalado Min-Max

```python
X_mm = minmax.fit_transform(X)
```

#### Escalado Robusto

```python
X_r = robust.fit_transform(X)
```

## 5) Reducción de la dimensionalidad (PCA)

La técnica de aprendizaje automático que primero vamos a implementar, al igual que en la memoria, es la reducción de la dimensionalidad aplicando el algoritmo (PCA). Para seleccionar el número de componentes vamos a establecer un mínimo del 70% de varianza explicada y posteriormente aplicar el criterio del codo para decidir el número final.

```python
#Importamos librería
from sklearn.decomposition import PCA

pca = PCA(n_components=0.7) #70% de la varianza explicada como mínimo
```

#### PCA - Escalado Estandar

```python
st_pca = pca.fit_transform(X_st)
print(f'Número de componentes (Standard): {st_pca.shape[1]}')
```

Necesitamos 5 variables para explicar al menos el 70% de la varianza.

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot - Criterio del Codo')
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.show()

print(pca.explained_variance_ratio_)
```

Para el escalado estandar, el número de componentes óptimo atendiendo al criterio del codo parece 3, aunque expicaríamos un porcentaje pequeño de la varianza.

#### PCA - Escalado Min-Max

```python
X_mm_pca = pca.fit_transform(X_mm)
print(f'Número de componentes (Min-Max): {X_mm_pca.shape[1]}')
```

En este caso, necesitamos el mismo número de variables para explicar el mismo porcentaje de varianza.

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot - Criterio del Codo')
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.show()

print(pca.explained_variance_ratio_)
```

De nuevo, atendiendo al criterio del codo, 3 parece el número adecuado de componentes, aunque en este caso observamos como la segunda componente explica más varianza.

#### PCA - Escalado Robusto

```python
X_r_pca = pca.fit_transform(X_r)
print(f'Número de componentes (Robust): {X_r_pca.shape[1]}')
```

Necesitamos una variable más que en el caso de los otros escalados.

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()

print(pca.explained_variance_ratio_)
```

La forma del gráfico nos hace dudar sobre el número correcto de componentes según este criterio ya que no existe un "codo" bien diferenciado.

### Selección final del modelo

Finalmente, consideramos que con el escalado min-max obtenemos una transformación que conserva un porcentaje mayor de varianza. En cuanto al número de componentes, por el criterio del codo, 3 sería el número adecuado explicando en torno al 60% de la varianza.

```python
final_pca = PCA(n_components=3)
X_final_pca = final_pca.fit_transform(X_mm)
```

## 6) Clustering

En esta sección, vamos a implementar los distintos algoritmos de clustering estudiados en la memoria. Estos son: 
- Clustering Jerárquico: aglomerativo empleando distancia mínima, máxima y media.
- Clustering Particional: K-means, K-medoids
- Clustering Basado en Densidad: DBSCAN

En general, el clustering funciona mejor cuando trabajamos con datasets con un número reducido de variables. Además, al trabajar con distancias, para su correcta implementación es necesario que las variables estén en un rango similar, es decir, que estén escaladas. Por estos motivos, y sin pérdida de generalidad, para la implementación y evaluación de estos modelos, vamos a utilizar el dataset con la transformación tras PCA con escalado estandar.

### 6.1) Clustering Jerárquico - Aglomerativo

Comenzamos con este tipo de clustering ya que no conocemos cúal puede ser el número óptimo de clusters. Para decidir este valor utilizaremos un dendograma. Vamos a distinguir distintos modelos en función del criterio de aglomeración: distancia máxima, mínima, media y de Ward; y a su vez de las distintas distacias estudiadas: euclídea, Manhattan, Chebyshov y Mahalanobis.

#### Distancia Máxima - Euclídea

```python
#Importamos las librerías
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering

plt.figure(figsize=(12, 7))  
plt.title("Dendrograma - Distancia Euclídea Máxima")  
dend = shc.dendrogram(shc.linkage(st_pca, method='complete', metric='euclidean')) #Distancia Euclídea Máxima
```

Distinguimos 5 clusters distintos, con clusters bastante bien diferenciados.

```python
agg_max_eu = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')  
agg_max_eu.fit(st_pca)
```

#### Distancia Máxima - Manhattan

```python
plt.figure(figsize=(12, 7))  
plt.title("Dendrograma - Distancia Manhattan Máxima")  
dend = shc.dendrogram(shc.linkage(st_pca, method='complete', metric='cityblock')) #Distancia Manhattan Máxima
```

De nuevo obtenemos 5 clusters, aunque la agrupación está mucho más descompensada.

```python
agg_max_man = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='complete')  
agg_max_man.fit(st_pca)
```

#### Distancia Máxima - Chebyshov

```python
plt.figure(figsize=(12, 7))  
plt.title("Dendrograma - Distancia Chebyshov Máxima")  
dend = shc.dendrogram(shc.linkage(st_pca, method='complete', metric='chebyshev')) #Distancia Chebyshov Máxima
```

### Evaluación

Vamos a emplear distintas métricas estudiadas para determinar que modelo y por tanto, qué número de clusters, se ajusta mejor a los datos.

```python
#Importamos las métricas
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

#Diccionario de modelos
Modelos = {
    'Max_Eu': agg_max_eu,
    'Max_Man': agg_max_man,
    'Max_Ch': agg_max_ch,
    'Max_Mah': agg_max_mah,
    'Min_Eu': agg_min_eu,
    'Min_Man': agg_min_man,
    'Min_Ch': agg_min_ch,
    'Avg_Eu': agg_av_eu,
    'Avg_Man': agg_av_man,
    'Avg_Ch': agg_av_ch,
    'Avg_Mah': agg_av_mah,
    'Ward': agg_ward,
}

#Dataframe de metricas
agg_metrics = pd.DataFrame(columns=['Modelo', 'Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])

for nom, mod in Modelos.items():
    agg_metrics = agg_metrics.append({
        'Modelo': nom,
        'Silhouette': silhouette_score(st_pca, mod.labels_),
        'Calinski-Harabasz': calinski_harabasz_score(st_pca, mod.labels_),
        'Davies-Bouldin': davies_bouldin_score(st_pca, mod.labels_)
    }, ignore_index=True)
    
agg_metrics
```

Recordemos que:
- El valor de Silhouette varía entre -1 y 1, de tal forma que un valor cercano a 1 indica que el punto está asignado correctamente, un valor de 0 indicaría indiferencia y un valor negativo indicaría que la observación estaría mejor agrupada en otro cluster.
- El coeficiente de Calinski-Harabasz mide la dispersión intra cluster y la dispersión entre los distintos clusters, de tal forma que un valor alto de este coeficiente indica una mejor calidad de clustering. 
- El coeficiente de Davies-Bouldin evalúa la compacidad y separación de los clusters, de tal forma que un valor bajo de esta métrica indicaría calidad en el agrupamiento.

Atendiendo a estas métricas, observamos que los modelos que emplearon el criterio de distancia mínima, obtienen muy buenos valores en las métricas de Silhouette y Davies-Bouldin, pero no obtendrían buenos resultados en la de Calinski-Harabasz. Observando sus dendogramas, está claro que estos modelos están muy afectadospor la presencia de uno o varios outliers que se diferencian enormemente del resto de valores.

Por otra parte el modelo de Ward es el que mejor valor obtiene en la métrica de Calinski-Harabasz, representando esta el ratio entre inercia y separación, además obtiene valores relativamente buenos de Silhouette. Observando el dendograma concluimos que esta división es la que mejor se ajusta a los datos y que por tanto, 2 es el número correcto de conglomerados.

### 6.2) Clustering Particional

Ahora, pasaremos a implementar los modelos de clustering particional estudiados, es decir, el K-Means y el K-Medoids. Aunque gracias al anterior análisis, hemos determinado que 2 es el número correcto de clusters evaluaremos un rango de posibles valores.

#### K-Means

```python
#Importamos las librerías
from sklearn.cluster import KMeans

#Rango de clusters
range_n_clusters = [2, 3, 4, 5]

#Dataframe de métricas
kmeans_metrics = pd.DataFrame(columns=['n', 'Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])

for num_clusters in range_n_clusters:
    
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50, random_state=23)
    kmeans.fit(st_pca)
  
    kmeans_metrics = kmeans_metrics.append({
        'n': str(num_clusters),
        'Silhouette': silhouette_score(st_pca, kmeans.labels_),
        'Calinski-Harabasz': calinski_harabasz_score(st_pca, kmeans.labels_),
        'Davies-Bouldin': davies_bouldin_score(st_pca, kmeans.labels_)
    }, ignore_index=True)
    
kmeans_metrics
```

### 6.3) Clustering Basado en Densidad - DBSCAN

Implementemos el último tipo de clustering. Probaremos distintas distancias, así como diferentes valores para los hiper-parámetros de radio y puntos mínimos.

```python
#Importamos las librerías
from sklearn.cluster import DBSCAN

#Rengos de valores
rad = [0.5, 0.75, 1]
MinP = [2,3]
met = ['euclidean', 'manhattan', 'chebyshev']

#Dataframe de métricas
DBSCAN_metrics = pd.DataFrame(columns=['Modelo', 'Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])


for r in rad:
    for p in MinP:
        for m in met:
            mod = DBSCAN(eps=r, min_samples=p, metric=m).fit(st_pca)
            
            DBSCAN_metrics = DBSCAN_metrics.append({
                'Modelo': f'r: {r}, MinP: {p}, {m}',
                'Silhouette': silhouette_score(st_pca, mod.labels_),
                'Calinski-Harabasz': calinski_harabasz_score(st_pca, mod.labels_),
                'Davies-Bouldin': davies_bouldin_score(st_pca, mod.labels_)
            }, ignore_index=True)
            
DBSCAN_metrics
```

Los resultados en general son malos ya que en la mayoría de casos obtenemos valores negativos de silhouette. Los mejores resultados se obtienen con la distancia de Chebysov y MinP=3. Vamos a ver si podemos mejorar estos resultados.

```python
#Rengos de valores
rad = [0.75, 1,1.5,2]
MinP = [3,4,5]

#Dataframe de métricas
DBSCAN_metrics2 = pd.DataFrame(columns=['Modelo', 'Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])

for r in rad:
    for p in MinP:
            mod = DBSCAN(eps=r, min_samples=p, metric='chebyshev').fit(st_pca)
            
            DBSCAN_metrics2 = DBSCAN_metrics2.append({
                'Modelo': f'r: {r}, MinP: {p}, chebyshev',
                'Silhouette': silhouette_score(st_pca, mod.labels_),
                'Calinski-Harabasz': calinski_harabasz_score(st_pca, mod.labels_),
                'Davies-Bouldin': davies_bouldin_score(st_pca, mod.labels_)
            }, ignore_index=True)
            
DBSCAN_metrics2
```

Observamos como a medida que aumentamos el tamaño del radio, el valor de silhouette también aumenta y el índice de Davies-Bouldin disminuye. Por otra parte, el índice Calinski-Harabasz aumenta al principio pero después disminuye. Además si observamos los valores de las métricas en las últimas filas vemos que coinciden con los modelos aglomerativos de distancia mínima que se veían muy afectados por los outliers.

Teniendo en cuenta que el algoritmo asigna el valor -1 al ruido (lo considera un cluster aparte), estudiemos algunos modelos 

```python
DB1 = DBSCAN(eps=1, min_samples=4, metric='chebyshev').fit(st_pca)

unique, counts = np.unique(DB1.labels_, return_counts=True)
print('r: 1, MinP: 4, chebyshev')
print('========================')
for i in unique:
    print(f'Cluster: {unique[i]}, Recuento: {counts[i]}')
```

```python
DB1 = DBSCAN(eps=2, min_samples=3, metric='chebyshev').fit(st_pca)

unique, counts = np.unique(DB1.labels_, return_counts=True)
print('r: 2, MinP: 3, chebyshev')
print('========================')
for i in unique:
    print(f'Cluster: {unique[i]}, Recuento: {counts[i]}')
```

No estamos consiguiendo buenas agrupaciones con este método ya que resulta difícil determinar el valor de los hiper-parámetros. En los modelos estudiados, solo distinguimos un único cluster y ruido.

Evaluamos los modelos mediante validación cruzada

```python
#Importamos las librerías
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, balanced_accuracy_score

#Definimos la métrica para la validación cruzada
def cross_validate_metrics(X, y):
    
    #Diccionario con todos los modelos
    Clas_Modelos = {
        'Reg_Log': LogisticRegression(random_state=r),
        'SVM_rbf': SVC(random_state=r),
        'SVM_poly': SVC(kernel='poly', random_state=r),
        'SVM_sigmoid': SVC(kernel='sigmoid', random_state=r),
        'SVM_linear': LinearSVC(random_state=r),
    }
    
    #Dataframe de métricas
    clas_metrics = pd.DataFrame(columns=['Modelo', 'Accuracy', 'F1', 'Recall', 'Precision', 'Balanced_Accuracy'])
    
    #Entrenamos y evaluamos cada modelo mediante validación cruzada estratificada (por defecto) con 10 divisiones
    for nom, mod in Clas_Modelos.items():
        results = cross_validate(mod, X, y, cv=10, scoring=(
            ['accuracy', 'f1', 'recall', 'precision', 'balanced_accuracy']))
    
        #Añadimos al dataframe de métricas el nombre y la media de cada métrica en las 10 evaluaciones para cada modelo 
        clas_metrics = clas_metrics.append({
            'Modelo': nom,
            'Accuracy': results['test_accuracy'].mean(),
            'F1': results['test_f1'].mean(),
            'Recall': results['test_recall'].mean(),
            'Precision':results['test_precision'].mean(),
            'Balanced_Accuracy':results['test_balanced_accuracy'].mean(),
        }, ignore_index=True)
        
    #Rediseñamos el dataframe 
    clas_metrics = clas_metrics.round(3).sort_values("F1", ascending=False)
    clas_metrics.index =clas_metrics['Modelo']
    clas_metrics.drop(['Modelo'], axis=1, inplace=True)
    return clas_metrics
```

```python
clas_metrics = cross_validate_metrics(X_train, y_train)
clas_metrics
```

```python
#Representamos las distintas métricas en gráficos de barras
plt.subplots(figsize = (15,15))

plt.subplot(3,2,1)
sns.barplot(x= clas_metrics.index , y= clas_metrics['Accuracy'])
plt.title ('Exactitud',{'fontsize':15})

plt.subplot(3,2,2)
sns.barplot(x= clas_metrics.index , y= clas_metrics['F1'])
plt.title ('Coef. F1',{'fontsize':15})

plt.subplot(3,2,3)
sns.barplot(x= clas_metrics.index , y= clas_metrics['Recall'])
plt.title ('Sensibilidad',{'fontsize':15})

plt.subplot(3,2,4)
sns.barplot(x= clas_metrics.index , y= clas_metrics['Precision'])
plt.title ('Precisión',{'fontsize':15})

plt.subplot(3,2,5)
sns.barplot(x= clas_metrics.index , y= clas_metrics['Balanced_Accuracy'])
plt.title ('Exactitud Balanceada',{'fontsize':15})

plt.show()
```

### 7.1) Regresión logística

```python
#Importamos métricas
from sklearn.metrics import confusion_matrix, classification_report

#Definimos la función para la matriz de confusión
def matriz_confusion (a, b):
    cf_matrix = confusion_matrix(a, b)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in
                cf_matrix.flatten()]
    labels = [f'{v1}\n{v2}' for v1, v2 in
          zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
    plt.show()
```

```python
Reg_log = LogisticRegression(random_state=r)
Reg_log.fit(X_train, y_train)

#Valores respuesta predichos sobre el conjunto de test
y_pred_rl = Reg_log.predict(X_test)
matriz_confusion(y_test, y_pred_rl)
print(classification_report(y_test, y_pred_rl))
```

### 7.3) SVM con Kernel Polinómico

```python
SVM_poly = SVC(kernel='poly', random_state=r)
SVM_poly.fit(X_train, y_train)

#Valores respuesta predichos sobre el conjunto de test
y_pred_sp = SVM_poly.predict(X_test)
matriz_confusion(y_test, y_pred_sp)
print(classification_report(y_test, y_pred_sp))
```

### 7.4) SVM con Kernel RBF

```python
SVM_rbf = SVC(random_state=r)
SVM_rbf.fit(X_train, y_train)

#Valores respuesta predichos sobre el conjunto de test
y_pred_sr = SVM_rbf.predict(X_test)
matriz_confusion(y_test, y_pred_sr)
print(classification_report(y_test, y_pred_sr))
```

Además, para llevar a cabo un correcto análisis de regresión lineal múltiple, deberíamos estudiar que variables son significativas en cada caso y habría que estudiar los supuestos de homocedasticidad, independencia y normalidad, así como analizar posibles problemas de multicolinealidad.
